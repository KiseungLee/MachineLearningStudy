```python
# 다변수를 이용한 선형회귀를 배움.

# 일단 3강까지의 방법으로 한번 구현해보고
# ex) hypothesis = x1*w1 + x2*w2 + x3*w3 + b

# 다음엔 행렬을 이용해 깔끔하게 구현해보자.
# ex) hypothesis = WX+b
```

지금까진 y = f(x)의 1변수를 알아봤다.

이번강의에선 y = f(x1, x2, x3, ... xn) 다변수를 다루는 걸 배워봄





그냥 행렬을 이용해 차원을 높이면 된다.

일단 위는 1개 데이터에 대한 것이고

여러 데이터셋에 대해선

이렇게 해주면 된다.

그런데 이제 하다보면 하나의 데이터 인풋을 여러 신경망 병렬적으로 넣어야 되는 경우도 발생한다. 그럴땐 

위와 같이 하면 n개인풋 -> m개 아웃풋 이런식으로 구성 가능

```python
# lab04-1 
# 일단 이번랩에선 m개의 아웃풋을 만드는것 까진 안함.
# 이번 랩에선 일단 전 강의에서 하던 방식으로 다변수를 넣어보고
# matrix를 사용해서 간단하게 바꿔본다.
```

```python
# 기존 방식
import tensorflow.compat.v1 as tf

# 데이터 암거나 넣음
x1_data = [73., 93., 89., 73., 96.]
x2_data = [54., 34., 63., 77., 98.]
x3_data = [76., 45., 99., 56., 45.]
y_data = [67., 56., 44., 56., 55]

# placeholder 일단 딱 봐도 하드코딩 되고 잘못되고 있다는걸 느낄 수 있다.
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# 변수
w1 = tf.Variable(tf.random_normal([1]), name='weight1')
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')
b = tf.Variable(tf.random_normal([1]), name='bias')

# 가설
hypothesis = x1*w1 + x2*w2 + x3*w3 + b

# 비용함수
cost = tf.reduce_mean(tf.square(hypothesis - Y))
# 최적화
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# 실행
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], feed_dict={x1:x1_data, x2:x2_data, x3:x3_data, Y: y_data})
    if step%10 ==0:
        print(step, "Cost: ", cost_val, "\nPrediction: \n", hy_val)
```

```python
# 위의 잘못된 코드를 옳게 바꿔보자

# 기존 방식
import tensorflow.compat.v1 as tf

# 데이터 암거나 넣음
x_data = [[73., 93., 89., 73., 96.],[54., 34., 63., 77., 98.] ,[76., 45., 99., 56., 45.]]
y_data = [67., 56., 44., 56., 55]

# placeholder 일단 딱 봐도 하드코딩 되고 잘못되고 있다는걸 느낄 수 있다.
X = tf.placeholder(tf.float32, shape=[None, 3]) # none자리는 데이터 갯수 자린데 None으로 두면 유동적으로 갯수 조절 가능하다는 것. 뒤 3은 차원수
Y = tf.placeholder(tf.float32, shape=[None, 1])

# 변수
W = tf.Variable(tf.random_normal([3, 1]), name='weight1')
b = tf.Variable(tf.random_normal([1]), name='bias')

# 가설 , matmul은 matrix multiply 겠지
hypothesis = tf.matmul(X, W) + b

# 비용함수
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# 최적화
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# 실행
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], feed_dict={X:x_data, Y: y_data})
    if step%10 ==0:
        print(step, "Cost: ", cost_val, "\nPrediction: \n", hy_val)
```
